```python
# First let's import the packages we will use in this project
# You can do this all now or as you need them
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
# Read in the Data

df = pd.read_csv('Excel Files/movies.csv')
```

```python
# Let's look at the Data

df.head()
```

```python
# Let's see if there is any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print(f'{col} - {pct_missing}%')
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
# Data types for our Columns

df.dtypes
```

```python
df.dropna(inplace=True)
```

```python
# Create Correct Year Column

# There are some values with inconsistent syntax like they are without comma so it is hard to convert them
fal = df[df['released'].apply(lambda x:len(x.split(','))==2)==False]['released'].index

df[df['released'].apply(lambda x:len(x.split(','))==2)==False]['released']

def con(x):
    mon = [['january', 'february', 'march', 'april', 'may', 'june', 'july',
          'august', 'september', 'october', 'november', 'december']]
    x=x.split(',')
    if len(x) == 2:
        return x[1][1:5]
    else:
        x = x[0].split()
        for i in x:
            if i not in mon:
                return i

df['yearcorrect'] = df['released'].apply(lambda x: con(x))

```

```python
df.head()
```

```python
df = df.sort_values(by = ['gross'], inplace=False, ascending=False)
```

```python
# Drop any Duplicates
df['company'].drop_duplicates().sort_values(ascending=True)
```

```python
# Plotting Heatmap of correlation
corr = df.corr()
corr
```

```python
# High Correlation between Budget and Gross
plt.figure(figsize=(10,8))
sns.heatmap(corr,annot=True)
plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")
```

```python
# Scatter plot with budget vs gross
# plt.figure(figsize=(8,6))
plt.scatter(x='budget',y='gross',data=df)
plt.title('Budget vs Gross')
plt.xlabel('Budget for Film')
plt.ylabel('Gross Earnings')
```

```python
df.head()
```

```python
# Plot Budget vs Gross using Seaborn

sns.lmplot(x='budget',y='gross',data=df,line_kws={'color':'blue'})
```

```python
# looks at Company
df_num = df

for col_names in df_num:
    if(df_num[col_names].dtype == 'object'):
        df_num[col_names] = df_num[col_names].astype('category')
        df_num[col_names] = df_num[col_names].cat.codes
        
df_num.head()
```

```python
plt.figure(figsize=(14,10))
sns.heatmap(df_num.corr(),annot=True)
plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")
```

```python
df_num.corr()
```

```python
correlation_matrix = df_num.corr()

corr_pairs = correlation_matrix.unstack()

corr_pairs
```
